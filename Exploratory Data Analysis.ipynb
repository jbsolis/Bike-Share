```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from biokit.viz import corrplot
```

```python
data = pd.read_csv("train.csv", header = 0)
data['month'] = pd.DatetimeIndex(data.datetime).month
data['dayofweek'] = pd.DatetimeIndex(data.datetime).dayofweek
data['hour'] = pd.DatetimeIndex(data.datetime).hour
data['year'] = pd.DatetimeIndex(data.datetime).year
data['day'] = pd.DatetimeIndex(data.datetime).day
data['windspeed'] = np.log10(data['windspeed'] + 1)
```

```python
from datetime import datetime, time
```

```python
def add_daylight(df):
    if (datetime(2011,3,13,2) < datetime(df['year'],df['month'],df['day'],df['hour']) < datetime(2011,11,6,2)):
        return 1
    elif (datetime(2012,3,11,2) < datetime(df['year'],df['month'],df['day'],df['hour']) < datetime(2012,11,4,2)):
        return 1
    else:
        return 0
```

```python
data['daylightsavings'] = data.apply(add_daylight, axis = 1)
```

```python
plt.hist(data['daylightsavings'], bins =3)
```

```python
c = corrplot.Corrplot(data)
c.plot()
```

```python
for month in range(1,len(data)):
    if data['month'][month] <= 3:
        data['quarter'] = 1
    elif data['month'][month] >=10 :
        data['quarter'] = 4
    elif data['month'][month] >= 4 & data['month'][month] <= 6:
        data['quarter'] = 2
    else:
        data['quarter'] = 3
```
